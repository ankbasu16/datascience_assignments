```python
from scipy.stats import shapiro

data = [0.909, 2.817, 0.121,1.456, 8, 2.00, 0.500, 1.478, 1.637, 6]

stats,p = shapiro(data)
print('stats = %.3f , p=%.4f'%(stats,p))

if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

```python
actual_array =np.array(df["age"])
std=np.std(actual_array)
mean  = np.mean(actual_array)
#mean = df['age'].mean()
print('Mean = %f '%mean)
print('Standard deviation = %f '%std)
left_r = 0
right_r = 0
arr=[1,2]

for i in arr :
    left_r = mean-(i)*std
    right_r=mean+(i)*std
    df1=df[(df["age"]>=left_r) & (df["age"]<=right_r)]
    mod_array = np.array(df1["age"])
    #print(np.count_nonzero(mod_array))
    print('Percentage of values in {} std {}'.format(i,(np.count_nonzero(mod_array)/(np.count_nonzero(actual_array)))))
          


    
```
