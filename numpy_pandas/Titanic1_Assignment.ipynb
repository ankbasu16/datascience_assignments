```python
import pandas as pd
import numpy as np

print("##Read data and describe it")
df = pd.read_csv(r'C:\sample_excel\titanic_dataset.csv',dtype={'fare':str})
print(df.describe(include='all'))

print()
print("##columns with missing data")
df.columns[df.isna().any()].tolist()

print()
print("##Print shape of dataset")
print(df.shape)

print()
print("##drop columns with more than 25% missing data")
df.dropna(thresh=int(.75*df.shape[0]),axis=1,inplace=True)
print()

dataTypes = df.dtypes
 
print('##Data type of each column of Dataframe :')
print(dataTypes)

print()
print('##Converting price to numeric :')
df['fare'] = df['fare'].str.replace(',','').str.replace('$','').astype('float')
print(df['fare'])

print()
print('##Columns with still missing values and their count')
print(df.isna().sum())

print()
print('##fill na with mean for fare and age column column.')
#print(df.fare.fillna(df.fare.mean()))
df.fare.fillna(df.fare.mean(),inplace=True)
df.age.fillna(df.age.mean(),inplace=True)

print()
print('##drop na embarked')
#print(df['embarked'])
df.dropna(subset=['embarked'], inplace=True)
print(df.isna().sum())

print()
print('##dump the dataframe to a csv file - titanic_filtered.csv ')
df.to_csv(r'C:\Users\arkav\Desktop\Training\titanic_filtered.csv.csv')


print()
print('##for surviced column replace 0 with D and 1 with A ')
df['survived'].replace(to_replace=0, 
                 value ="D",inplace=True)
df['survived'].replace(to_replace=1, 
                 value ="A",inplace=True)
print(df.head(5))

print()
print('##find the frequency of different values in survived column ')
count = df['survived'].value_counts() 
print(count)

print()
print('##group by gender  and see the counts in each category')
grp=df.groupby(by="gender")
print(grp.gender.count())

print()
print('##group by  survived and see the counts in each category')
grp=df.groupby(by="survived")
print(grp.survived.count())


print()
print('##find different pclass and no of people in each class')
print(df.pclass.unique())
print('##no of people in each class ')
pclass_count = df['pclass'].value_counts() 
print(pclass_count)

##find top 5 people with highest values of age. Count no of male and females in the top 5
print()
print('##find top 5 people with highest values of age')
print(df.sort_values(by="age",ascending=False).head(5).name)
print()
print('##Count no of male and females in the top 5')
print(df.sort_values(by="age",ascending=False).head(5).groupby(by="gender").gender.count())

print()
print("##find max age male and female who survived")
#print(df.sort_values(by="age",ascending=False).head(5))
print(df.sort_values(by="age",ascending=False).groupby(by="gender").age.max())

print()
print("##get average age by gender")
print()
grp=df.groupby(by="gender")
print(grp.age.mean())

print()
print("##get average age by people survived vs not-survived")
print()
grp=df.groupby(by="survived")
print(grp.age.mean())





```
